```python
# 전국 & 서울 + 각 구별 인구 & 수원 & 창원 & 진주 인구 데이터 ( 2019년 )
df_pop = pd.read_csv("data/행정구역_시군구_별_주민등록세대수.csv", encoding = 'ansi')
df_pop.columns = ['city','pop']
df_pop.set_index('city', inplace=True)
pop_seoul = df_pop.iloc[0:27]
pop_else = df_pop.loc[['수원시', '창원시', '진주시']]
df_pop = pop_seoul.append(pop_else)
display(df_pop)

df_pop.to_csv('output/2019년_전국+서울전체+서울구별+수원+창원+진주_인구.csv', encoding='utf-8')
```

```python
# 서울 관할구 연령별 디테일 인구특성 분포 (2021 2분기)

df_seoul = pd.read_csv("data/연령별_구별_서울시_주민등록인구_2분기_2021.csv", encoding = 'ansi')
df_seoul.drop(['연도','분기'],axis=1,inplace=True)
#print(df_seoul.dtypes)
df_seoul.replace(' ', 0, inplace=True)
df_seoul['연령90_94세'] = df_seoul['연령90_94세'].astype(int)
df_seoul['연령95_100세'] = pd.to_numeric(df_seoul['연령95_100세'])
df_seoul['연령_100세이상'] =pd.to_numeric(df_seoul['연령_100세이상'])
#display(df_seoul.info())
#print(df_seoul.isblank().sum(axis=1))
df_seoul['10세 미만'] = df_seoul['연령0_4세']+df_seoul['연령5_9세']
df_seoul['10대'] = df_seoul['연령10_14세']+df_seoul['연령15_19세']
df_seoul['20대'] = df_seoul['연령20_24세']+df_seoul['연령25_29세']
df_seoul['30대'] = df_seoul['연령30_34세']+df_seoul['연령35_39세']
df_seoul['40대'] = df_seoul['연령40_44세']+df_seoul['연령45_49세']
df_seoul['50대'] = df_seoul['연령50_54세']+df_seoul['연령55_59세']
df_seoul['60대'] = df_seoul['연령60_64세']+df_seoul['연령65_69세']
df_seoul['70세 이상'] = df_seoul['연령70_74세']+df_seoul['연령75_79세']+df_seoul['연령80_84세']+df_seoul['연령85_89세']+df_seoul['연령90_94세']+df_seoul['연령95_100세']+df_seoul['연령_100세이상']

df_seoul.drop(df_seoul.columns[3:24],axis=1,inplace=True)
df_seoul.drop(df_seoul.index[38],axis=0,inplace=True)
display(df_seoul)

df_seoul.to_csv('output/2021년_서울_관할지역_연령별_내외국인분포.csv', encoding='utf-8')
```
