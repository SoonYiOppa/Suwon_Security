```python
one_person = pd.read_csv('data/33.4개시도_1인가구수.csv', encoding='utf-8')
one_person.columns = ['지역', '관할구','성별','연령','시점','1인가구','주택_계', '단독주택','아파트','연립주택','다세대주택','비거주용건물내주택','주택외거처']

one_person = one_person.loc[one_person['연령']=='합계']
display(one_person)
one_person.replace('X', 0, inplace=True)

one_person['아파트'] = one_person['아파트'].astype(int)
one_person['연립주택'] = one_person['연립주택'].astype(int)
one_person['다세대주택'] = one_person['다세대주택'].astype(int)
one_person['비거주용건물내주택'] = one_person['비거주용건물내주택'].astype(int)
one_person['주택외거처'] = one_person['주택외거처'].astype(int)
display(one_person.info())

one_person.drop('연령', axis=1, inplace=True)

total_group = one_person.groupby(['지역','관할구','시점']).sum()
display(total_group[10:32])
total_group.to_csv('output/4개시도_연도별_1인가구+주택현황.csv',encoding='utf-8')
```
