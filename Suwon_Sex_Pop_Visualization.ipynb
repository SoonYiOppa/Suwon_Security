```python
import pandas as pd
import folium
from folium import plugins
import json
import geopandas as gpd
```

```python
### [[수원시]] _ GeoJson
## MultiLineString
suwon_road = json.load(open('data/geojson/28.수원시_도로명주소(도로).geojson', encoding='utf-8'))

##MultiPolygon
#격자 기반
suwon_grid = json.load(open('data/geojson/32.수원시_표준격자(250M_x_250M).geojson', encoding='utf-8'))
suwon_grid_report = json.load(open('data/geojson/2.수원시_112신고(격자별 신고횟수).geojson', encoding='utf-8'))
suwon_grid_SexAge = json.load(open('data/geojson/30.수원시_성연령별_거주인구격자(250M_X_250M).geojson', encoding='utf-8'))
#행정동
suwon_Area = json.load(open('data/geojson/31.수원시_행정경계(읍면동).geojson', encoding='utf-8'))
```

```python
#####################################
#행정동별_연령별 남/녀 거주인구 분포#
m_20g,m_30g,m_40g,m_50g,m_60g,m_70g,m_80g,m_90g,m_100g  = [],[],[],[],[],[],[],[],[]
w_20g,w_30g,w_40g,w_50g,w_60g,w_70g,w_80g,w_90g,w_100g  = [],[],[],[],[],[],[],[],[]
SexAge_group = ['m_20g_pop', 'w_20g_pop', 
                'm_30g_pop', 'w_30g_pop', 
                'm_40g_pop', 'w_40g_pop', 
                'm_50g_pop', 'w_50g_pop', 
                'm_60g_pop', 'w_60g_pop', 
                'm_70g_pop', 'w_70g_pop', 
                'm_80g_pop', 'w_80g_pop', 
                'm_90g_pop', 'w_90g_pop', 
                'm_100_pop', 'w_100g_pop']
grid_age = []
gid = []
#print(suwon_grid_SexAge['features'][0]['properties'][SexAge_group[0]])

for i in range(len(suwon_grid['features'])):
    gid.append(suwon_grid['features'][i]['properties']['gid'])
    age = []
    for k in range(len(suwon_grid_SexAge['features'][i]['properties']) - 1) :
        if suwon_grid_SexAge['features'][i]['properties'][SexAge_group[k]] == None :
            age.append(0)
        else :
            age.append(suwon_grid_SexAge['features'][i]['properties'][SexAge_group[k]])
        
    grid_age.append(age)
    
# print(len(gid))
# print(grid_age)

# 연령대별 남녀

for i in range(len(gid)):
    m_20g.append(grid_age[i][0])
    m_30g.append(grid_age[i][2])
    m_40g.append(grid_age[i][4])
    m_50g.append(grid_age[i][6])
    m_60g.append(grid_age[i][8])
    m_70g.append(grid_age[i][10])
    m_80g.append(grid_age[i][12])
    m_90g.append(grid_age[i][14])
    m_100g.append(grid_age[i][16])
    w_20g.append(grid_age[i][1])
    w_30g.append(grid_age[i][3])
    w_40g.append(grid_age[i][5])
    w_50g.append(grid_age[i][7])
    w_60g.append(grid_age[i][9])
    w_70g.append(grid_age[i][11])
    w_80g.append(grid_age[i][13])
    w_90g.append(grid_age[i][15])
    w_100g.append(grid_age[i][17])

 
    
m = []
w = []
# 전체 남녀
for i in range(len(gid)):
    m.append(m_20g[i]+m_30g[i]+m_40g[i]+m_50g[i]+m_60g[i]+m_70g[i]+m_80g[i]+m_90g[i]+m_100g[i])
    w.append(w_20g[i]+w_30g[i]+w_40g[i]+w_50g[i]+w_60g[i]+w_70g[i]+w_80g[i]+w_90g[i]+w_100g[i])


# print(len(gid))
# print(len(m_20g))
# print(len(m))

grid_20pop = pd.DataFrame({'gid':gid,
                           'male':m_20g,
                          'female':w_20g})  
grid_30pop = pd.DataFrame({'gid':gid,
                           'male':m_30g,
                          'female':w_30g})
grid_40pop = pd.DataFrame({'gid':gid,
                           'male':m_40g,
                          'female':w_40g})  
grid_50pop = pd.DataFrame({'gid':gid,
                           'male':m_50g,
                          'female':w_50g})  
grid_60pop = pd.DataFrame({'gid':gid,
                           'male':m_60g,
                          'female':w_60g})  
grid_70pop = pd.DataFrame({'gid':gid,
                           'male':m_70g,
                          'female':w_70g})  
grid_80pop = pd.DataFrame({'gid':gid,
                           'male':m_80g,
                          'female':w_80g})  
grid_90pop = pd.DataFrame({'gid':gid,
                           'male':m_90g,
                          'female':w_90g})  
grid_100pop = pd.DataFrame({'gid':gid,
                           'male':m_100g,
                          'female':w_100g})  

grid_male = pd.DataFrame({'gid':gid,
                           'pop':m})  
grid_female = pd.DataFrame({'gid':gid,
                           'pop':w})  
```

```python
## 20대
center = [37.280364359333866, 127.00776116595848] # 수원 가운데

suwon_20pop = folium.plugins.DualMap(
                                location=center,
                                tiles='cartodbpositron',
                                zoom_start=12)


grid_20male = folium.Choropleth(
                    geo_data = suwon_grid,
                    data = grid_20pop,
                    columns = ('gid', 'male'),
                    key_on = 'properties.gid',
                    line_weight = 0.3,
                    fill_color = 'Blues',
                    fill_opacity = 1, 
                    legend_name = '격자구역 별 20대 남성 거주인구',
                    threshold_scale=[0, 75, 150, 225, 300, 375, 450, 550 , 650, 790],
                    highlight=True,    
                ).add_to(suwon_20pop.m1)

# grid_rep_map.geojson.add_child(
#     folium.features.GeoJsonTooltip(['gid'], labels= False)
# )

grid_20female = folium.Choropleth(
                    geo_data = suwon_grid,
                    data = grid_20pop,
                    columns = ('gid', 'female'),
                    key_on = 'properties.gid',
                    line_weight = 0.3,
                    fill_color = 'YlOrRd',
                    fill_opacity = 1, 
                    legend_name = '격자구역 별 20대 여성 거주인구',
                    threshold_scale=[0, 75, 150, 225, 300, 375, 450, 550 , 650, 790],
                    highlight=True,    
                ).add_to(suwon_20pop.m2)

dong_area_map = folium.Choropleth(
                    geo_data = suwon_Area,
                    line_weight = 0.3,
                    line_color = 'black',
                    line_opacity = 1,
                    fill_opacity = 0, 
                    highlight = True,
                ).add_to(suwon_20pop)
dong_area_map.geojson.add_child(
    folium.features.GeoJsonTooltip(['ADM_DR_NM'],labels=False)
)

# folium.LayerControl().add_to(suwon_gmap)
display(suwon_20pop)

suwon_20pop.save('수원시_행정동_20대_성별분포.html')
```
